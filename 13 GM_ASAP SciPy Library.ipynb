QS.2. According to the U.S. census bureau, approximately 6% of all workers in jackson,Mississipi are uneployed.In conducting a random telephone survey in jackson,what is the probability of getting two or the fewer unemployed workers in a sample of 20?

```python
import scipy
import numpy
from scipy.stats import binom
##probability of unemployed in jackson =p=0.06 
##sample population = n= 20
##getting 2 0r fewer = K= 2
print(binom.cdf(k=2,n=20,p=0.06))
         ## OR
print(binom.cdf(2,20,0.06))
```

```python
p1=binom.pmf(2,20,0.06)
p1
```

```python
p2=binom.pmf(1,20,0.06)
p2
```

```python
p3=binom.pmf(0,20,0.06)
p3
```

```python
ans=p1+p2+p3
print(ans)
```
